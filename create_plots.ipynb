```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import functions
```

```python
list_of_functions = ["linear1_abrupt", "linear2_abrupt", "linear3_abrupt",
                    "nonlinear1_abrupt", "nonlinear2_abrupt", "nonlinear3_abrupt"]
```

```python
for name in list_of_functions:
#   load the data
    data = pd.read_csv("data/"+name, usecols = [0]).iloc[:,0].to_list()
    
#   plot the data with lines indicating change of the concepts
    plt.plot(data, color = "black")
    plt.vlines(x = [200,400,600], ymin = data.min(), ymax = data.max(), 
               linestyles = "dashed", color = "red")

    #saving the plot
    image_path = "plots/"+name+".png"
    plt.savefig(image_path)
    
    plt.show()
```
